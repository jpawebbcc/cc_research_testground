```python
import pandas as pd
import numpy as np
import time
import requests
```

Example query to look at Uniswap v2 and v3 subgraphs - it might be a good idea to split them as different exchanges for calculations.

You can use the playground to figure out queries and the copy that into the code below adjusting for different variables. the scheama is pretty helpful if your having trouble or its spitting errors just let me know and I can offer some help. 

Uniswap v2 Subgraph - https://thegraph.com/hosted-service/subgraph/uniswap/uniswap-v2

Uniswap v3 Subgraph - https://thegraph.com/hosted-service/subgraph/uniswap/uniswap-v3 

```python
#Uniswap v2
conid = pd.DataFrame()
txc = 45251176 # some starting txncount to start iterating down from
i = 0 
while True:
    s = len(conid)
    query = '''
{
    pairs(where: {txCount_lt: "'''+f'{txc}'+'''"}, orderBy: txCount, orderDirection: desc, first: 1000) {
    id
    token0 {
      symbol
      id
    }
    token1 {
      symbol
      id
    }
    txCount
    volumeToken0
    volumeToken1
    volumeUSD
    createdAtTimestamp
  }
}
'''
    request = requests.post('https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v2'
                                '',
                                json={'query':query})
    data = request.json()
    try:
        tdd = pd.json_normalize(data['data']['pairs'])
        conid = conid.append(tdd)
        txc = conid['txCount'].iloc[-1]
        i += 1
        if s == len(conid):
            break
    except Exception as e:
        print(e)
        time.sleep(5)
    if i > 500: 
        break
conid

```
